```python
import pandas as pd
import json
```

MATCH RESULTS

```python
with open('t20_json_files/t20_wc_match_results.json') as json_file:
    data = json.load(json_file)
df_match = pd.DataFrame(data[0]['matchSummary'])   
print(df_match.shape)
df_match.head()
```

```python
df_match.rename({'scorecard': 'match_id'}, axis = 1, inplace = True)
df_match.head()
```

```python
#creating match ids dictionary to link match results with batting summary

match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']

match_ids_dict    
```

BATTING SUMMARY

```python
with open('t20_json_files/t20_wc_batting_summary.json') as json_file:
    data = json.load(json_file)
    
    all_record = []
    
    for rec in data:
        all_record.extend(rec['battingSummary'])

df_batting = pd.DataFrame(all_record)
print(df_batting.shape)
df_batting.head()        
```

```python
df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x: "out" if len(x) > 0 else "not_out")
df_batting.head(15)
```

```python
df_batting.drop(columns=["dismissal"], inplace=True)
df_batting.head()
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head()
```

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
df_batting.head()
```

BOWLING SUMMARY

```python
with open('t20_json_files/t20_wc_bowling_summary.json') as json_file:
    data = json.load(json_file)
    
    all_record = []
    
    for rec in data:
        all_record.extend(rec['bowlingSummary'])    
        
df_bowling = pd.DataFrame(all_record)
print(df_bowling.shape)
df_bowling.head()        
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_ids_dict)
df_bowling.head()
```
